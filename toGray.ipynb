```python
import cv2
from PIL import Image
import numpy as np

file = 'Children_resize/1.jpg'
img = cv2.imread(file, cv2.IMREAD_COLOR)

imgPIL = Image.open(file)

kq = Image.new(imgPIL.mode, imgPIL.size)

width = kq.size[0]
height = kq.size[1]

for i in range(width):
    for j in range(height):
        R,G,B = imgPIL.getpixel(i,j)
        gray = np.uint8(0.2989 * R + 0.5870 * G + 0.1140 * B)
        kq.putpixel((i,j),gray)

ax = np.array(kq)
cv2.imshow(ax)
```
